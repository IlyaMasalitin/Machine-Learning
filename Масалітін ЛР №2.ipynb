```python
#9-13
import pandas as pd
import numpy as np
import seaborn as sns

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
tables = pd.read_html(url)
df = tables[2]  

print("Назви стовпців:")
print(df.columns)
print("\nПерші 5 рядків даних:")
print(df.head())

num_columns = df.shape[1]
print("\nКількість стовпців:", num_columns)

df.columns = ['Country', 'GDP_2024', 'WorldBank_2023', 'OON_2022', 'Other1', 'Other2', 'Other3']

print("\nПропущені значення до заповнення:")
print(df.isna().sum())

df.replace("—", np.nan, inplace=True)

numeric_columns = ['GDP_2024', 'WorldBank_2023', 'OON_2022']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

df.fillna(df.mean(numeric_only=True), inplace=True)

print("\nПропущені значення після заповнення:")
print(df.isna().sum())

print("\nПеревірка наявності дублікатів:")
print(df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("\nОписова статистика датасету:")
print(df.describe())


df['Difference_MVF_WB'] = df['GDP_2024'] - df['WorldBank_2023']
max_diff_country = df.loc[df['Difference_MVF_WB'].abs().idxmax()]
print("\nКраїна з найбільшою різницею між показниками:")
print(max_diff_country)


correlation_matrix = df[['GDP_2024', 'WorldBank_2023', 'OON_2022']].corr()
print("\nКореляційна матриця:")
print(correlation_matrix)


mean_values = df[['GDP_2024', 'WorldBank_2023', 'OON_2022']].mean()
print("\nСереднє значення для кожного стовпця:")
print(mean_values)


std_dev = df[['GDP_2024', 'WorldBank_2023', 'OON_2022']].std()
print("\nСтандартне відхилення для кожного стовпця:")
print(std_dev)
```

```python
#13-17
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
tables = pd.read_html(url)
df = tables[2]  

print("Назви стовпців:")
print(df.columns)
print("\nПерші 5 рядків даних:")
print(df.head())

df.columns = ['Country', 'GDP_2024', 'WorldBank_2023', 'OON_2022', 'Other1', 'Other2', 'Other3']

df.replace("—", np.nan, inplace=True)

numeric_columns = ['GDP_2024', 'WorldBank_2023', 'OON_2022']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

print("\nПропущені значення до заповнення:")
print(df.isna().sum())

df.fillna(df.mean(numeric_only=True), inplace=True)

print("\nПропущені значення після заповнення:")
print(df.isna().sum())

highest_gdp_country = df.loc[df['GDP_2024'].idxmax()]
lowest_gdp_country = df.loc[df['GDP_2024'].idxmin()]

highest_wb_country = df.loc[df['WorldBank_2023'].idxmax()]
lowest_wb_country = df.loc[df['WorldBank_2023'].idxmin()]

highest_oon_country = df.loc[df['OON_2022'].idxmax()]
lowest_oon_country = df.loc[df['OON_2022'].idxmin()]

print("\nКраїна з найвищим ВВП (GDP_2024):")
print(highest_gdp_country)

print("\nКраїна з найнижчим ВВП (GDP_2024):")
print(lowest_gdp_country)

print("\nКраїна з найвищим WorldBank_2023:")
print(highest_wb_country)

print("\nКраїна з найнижчим WorldBank_2023:")
print(lowest_wb_country)

print("\nКраїна з найвищим OON_2022:")
print(highest_oon_country)

print("\nКраїна з найнижчим OON_2022:")
print(lowest_oon_country)

plt.figure(figsize=(10, 6))
sns.histplot(df['GDP_2024'], kde=True, bins=30)
plt.title('Розподіл GDP_2024 серед країн')
plt.xlabel('GDP (млн. дол. США)')
plt.ylabel('Частота')
plt.show()

df['GDP_share'] = df['GDP_2024'] / df['GDP_2024'].sum()
df['WorldBank_share'] = df['WorldBank_2023'] / df['WorldBank_2023'].sum()
df['OON_share'] = df['OON_2022'] / df['OON_2022'].sum()

print("\nЧастка кожної країни у GDP_2024:")
print(df[['Country', 'GDP_share']].head())

print("\nЧастка кожної країни у WorldBank_2023:")
print(df[['Country', 'WorldBank_share']].head())

print("\nЧастка кожної країни у OON_2022:")
print(df[['Country', 'OON_share']].head())

plt.figure(figsize=(12, 8))
for country in df['Country'].head(10):  
    country_data = df[df['Country'] == country]
    plt.plot(['GDP_2024', 'WorldBank_2023', 'OON_2022'],
             country_data[['GDP_2024', 'WorldBank_2023', 'OON_2022']].values.flatten(),
             label=country)

plt.title('Зміни в показниках для кожної країни (топ 10)')
plt.xlabel('Показники')
plt.ylabel('Значення')
plt.legend()
plt.show()
```

```python
#14-22
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('D:\\titanic (2).csv', delimiter=',', header=0)

print("\nОписова статистика датасету:")
print(df.describe())

if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)

df_selected = df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']]
print("\nПерші 5 рядків відфільтрованого датасету:")
print(df_selected.head())

print("\nУнікальні значення стовпця 'Sex':")
print(df_selected['Sex'].unique())

df_selected.loc[:, 'Sex'] = df_selected['Sex'].map({'male': 0, 'female': 1})

print("\nПропущені значення після заміни статі:")
print(df_selected.isna().sum())

print("\nПерші 5 рядків відфільтрованого датасету:")
print(df_selected.head())

print("\nОстанні 5 рядків відфільтрованого датасету:")
print(df_selected.tail())

survival_by_sex = df_selected.groupby('Sex')['Survived'].mean() * 100
print("\nВідсоток виживання для кожної статі:")
print(survival_by_sex)

survival_by_class = df_selected.groupby('Pclass')['Survived'].mean() * 100
print("\nВідсоток виживання для кожного класу:")
print(survival_by_class)

mean_age_survived = df_selected[df_selected['Survived'] == 1]['Age'].mean()
mean_age_not_survived = df_selected[df_selected['Survived'] == 0]['Age'].mean()
print(f"\nСередній вік тих, хто вижив: {mean_age_survived:.2f}")
print(f"Середній вік тих, хто не вижив: {mean_age_not_survived:.2f}")

df_selected = df_selected.copy()  
df_selected.loc[:, 'Fare_group'] = pd.cut(df_selected['Fare'], bins=[0, 10, 50, 100, 500], labels=['<10', '10-50', '50-100', '100+'], include_lowest=True)
survival_by_fare_group = df_selected.groupby('Fare_group', observed=False)['Survived'].mean() * 100
print("\nРівень виживання для кожної групи тарифів:")
print(survival_by_fare_group)

average_fare_by_class = df_selected.groupby('Pclass')['Fare'].mean()
print("\nСередній тариф для кожного класу:")
print(average_fare_by_class)

average_age_by_class = df_selected.groupby('Pclass')['Age'].mean()
print("\nСередній вік пасажирів для кожного класу:")
print(average_age_by_class)

plt.figure(figsize=(12, 6))
for survived in [0, 1]:
    subset = df_selected[df_selected['Survived'] == survived]
    plt.hist(subset['Age'].dropna(), bins=30, alpha=0.5, label=f"Survived={survived}")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Розподіл віку для тих, хто вижив і тих, хто не вижив')
plt.legend()
plt.show()

survival_by_sex_class = df_selected.groupby(['Sex', 'Pclass'])['Survived'].mean() * 100
print("\nВідсоток виживання для кожної комбінації статі та класу:")
print(survival_by_sex_class)

numeric_df_selected = df_selected.select_dtypes(include=[np.number])
correlation_matrix = numeric_df_selected.corr()
print("\nКореляційна матриця:")
print(correlation_matrix)
```
