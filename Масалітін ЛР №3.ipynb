```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
california_housing = fetch_california_housing()
data = california_housing.data
feature_names = california_housing.feature_names
target = california_housing.target
df = pd.DataFrame(data, columns=feature_names)
df['Target'] = target
print(df.head())
```

```python
print("\nОписова статистика:")
print(df.describe())
print("\nПропущені значення:")
print(df.isnull().sum())
print("\nДублікати:")
print(df.duplicated().sum())
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = df[feature_names]
y = df['Target']
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(f"Розмір тренувальної вибірки X: {X_train.shape}")
print(f"Розмір тестової вибірки X: {X_test.shape}")
print(f"Розмір тренувальної вибірки y: {y_train.shape}")
print(f"Розмір тестової вибірки y: {y_test.shape}")

print("\nПерші 5 рядків тренувальної вибірки X:")
print(X_train[:5])

print("\nПерші 5 значень тренувальної вибірки y:")
print(y_train[:5])

print("\nПерші 5 рядків тестової вибірки X:")
print(X_test[:5])

print("\nПерші 5 значень тестової вибірки y:")
print(y_test[:5])
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
    'Ridge': Ridge()
}
param_grids = {
    'Random Forest': {
        'n_estimators': [10, 50, 100],
        'max_depth': [None, 10, 20, 30]
    },
    'Ridge': {
        'alpha': [0.1, 1, 10, 100]
    }
}

results = {}

for model_name, model in models.items():
    print(f"\nТренування моделі {model_name}...")
    if model_name in param_grids:
        grid_search = GridSearchCV(model, param_grids[model_name], cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
    else:
        best_model = model.fit(X_train, y_train)
    
    y_pred = best_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results[model_name] = {
        'Best Model': best_model,
        'MSE': mse,
        'R2': r2
    }
    
    print(f"{model_name} - Best MSE: {mse:.2f}, R^2: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt

for model_name, result in results.items():
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, result['Best Model'].predict(X_test), alpha=0.5)
    plt.xlabel('True Values')
    plt.ylabel('Predictions')
    plt.title(f'{model_name} - True Values vs Predictions')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
    plt.show()

for model_name, result in results.items():
    print(f"\n{model_name} - Перші 5 справжніх і прогнозованих значень:")
    predictions = result['Best Model'].predict(X_test)
    comparison_df = pd.DataFrame({'True Values': y_test, 'Predictions': predictions})
    print(comparison_df.head())
```
